##Matrix Multiplications
In machine learning, matrix multiplication is one of the most common tensor operations

There are two rules our tensors(or matrices) need to fullfill if we're going to matrix mutiply them;

1. The inner dimenstions must match
2. the resulting matrix has the shape of the outer dimensions

```python
#matrix multification tensorflow
print(tensor)
tf.matmul(tensor,tensor)
```

```python
#matrix multiplication with python operator "@"
tensor @ tensor
```

```python
tensor.shape
```

```python
#create  a tensor (3,2) tensor
X = tf.constant([[1,2],[3,4],[5,6]])
#Create another(3,2) tensor
Y = tf.constant([[7,8],[9,10],[11,12]])
X,Y
```

```python
#try to matrix multiply tensors of same shape
#X @ Y #error follow the 2 rules
#tf.matmul(X,Y) #error follow the 2 rules
```

##Find the positional minimum and maximum

```python
#create a new tensor for finding positional minimum and maximum
tf.random.set_seed(42)
F=tf.random.uniform(shape=[50])
F
```

```python
#find the positional maximum
tf.argmax(F)
```

```python
#index on our largest value postion
F[tf.argmax(F)] #grabbing value of argmax
```

```python
#find the max value of
tf.reduce_max(F) #grabbing value of reduce_max
```

```python
#check for equality
F[tf.argmax(F)] == tf.reduce_max(F)
```

```python
#find the positional minimum
tf.argmin(F)
```

```python
#find the minimum using the postional minimum index
F[tf.argmin(F)] #grabbing value of argmin
```
